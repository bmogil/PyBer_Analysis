```python
# Import dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
```

```python
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'ro')
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare(%)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title. 
plt.title("Pyber Fare by Month")

# Add the legend.
plt.legend()
```

#### Bar Graphs

##### Vertical Bar Graph

```python
plt.bar(x_axis, y_axis)
```

```python
# Annotating Bar Graph

# Create plot
plt.bar(x_axis, y_axis, color='green', label='Boston')

# Create labels for the x and y axes
plt.xlabel('Date')
plt.ylabel('Fare($)')

# Create a title
plt.title('PyBer Fare by Month')

# Add the legend
plt.legend()
```

##### Horizontal Bar Graph

```python
plt.barh(x_axis, y_axis, label='Boston', color='green')

plt.title('PyBer Fare by Month')

plt.xlabel('Date')
plt.ylabel('Fare($)')

plt.legend()
```

```python
# If we want to invert the y-axis values so that Jan is at the top: 

plt.barh(x_axis, y_axis, color='green', label='Boston')

# The gca() method stands for 'get current axes' ... so we use it to specify the current axes and then further specify which of the current axes we want ... and what we want to do to it with the 'invery_yaxis()'
plt.gca().invert_yaxis()
```

```python
# If we want the above graph but the data to be on the opposite axes:
plt.barh(y_axis, x_axis, label='Boston', color='green')

plt.title('PyBer Fare by Month')

plt.xlabel('Date')
plt.ylabel('Fare($)')

plt.legend()

# For this data ... this is stupid ... but good to know how to invert the axes.
```

#### Scatter Plot

```python
plt.scatter(y_axis, x_axis, color='red', label='Chicago')

plt.legend()

plt.title('PyBer Fare by Month')

plt.xlabel('Date')
plt.ylabel('Fare ($)')

plt.xlim(0, 50)

plt.gca().invert_yaxis()
```

#### Bubble Chart

```python
# We can transform the above scatter plot into a bubble chart by using the 's' parameter.
plt.scatter(x_axis, y_axis, s=y_axis, color='red', label='Chicago')
```

```python
# Since some of the markers are too small ... we'll use a for loop to iterate through each y_axis value and multiply them by 3 and then an empty list to store the new values.

# Defining empty list
y_axis_larger = []

# for loop
for data in y_axis:
    y_axis_larger.append(data*3)
    
# Then we'll simply use this new list as the y_axis value for the 's' param. when creating the chart.
plt.scatter(x_axis, y_axis, s=y_axis_larger, color='red', label='Chicago')
```

```python
# We can also use list comp. to achieve the same feat. as above, instead of a for loop --> by using list comp. within the plt.scatter() method for the 's' param.
plt.scatter(x_axis, y_axis, color='red', label='Chicago', s=[i*3 for i in y_axis])
```

```python
# 5.1.7 Skill Drill

plt.scatter(y_axis, x_axis, color='skyblue', label='Boston', s=[i*5 for i in y_axis], edgecolor='black', alpha=0.8, linewidths=2)

plt.legend()

plt.title('Pyber Fare by Month')

plt.xlim(0, 50)

plt.gca().invert_yaxis()
```

#### Pie Chart

```python
plt.pie(y_axis, labels=x_axis)
```

```python
# Adding a percentage for each wedge

plt.pie(y_axis, labels=x_axis, autopct='%.1f%%')
```

```python
# Utilizing 'explode' parameter

explode_values = [0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0]

plt.subplots(figsize=(8,8))

plt.pie(y_axis, labels=x_axis, autopct='%.1f%%', explode=explode_values)
```

```python
# Customizing each wedge color

colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

explode_values = [0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0]

plt.subplots(figsize=(8,8))

plt.pie(y_axis, labels=x_axis, autopct='%.1f%%', explode=explode_values, colors=colors, shadow=True, startangle=84)

plt.gca().invert_yaxis()
```

#### Error Bars

```python
import statistics
```

```python
stdev = statistics.stdev(y_axis)

plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)

plt.grid()
```
